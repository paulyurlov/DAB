```python
def numeric_description(df, numeric):
    df_res = df[numeric].describe().transpose()
    mdn = []
    nulls = []
    for name in numeric:
        mdn.append(np.median(df[name].values))
        nulls.append(df[name].isnull().sum() / df[name].shape[0])
    df_res['median'] = mdn
    df_res['part_missing'] = nulls
    return df_res
```

```python
from scipy.stats import mode
def categorical_description(df, numeric):
    df_res = pd.DataFrame()
    mdn = []
    nulls = []
    for name in numeric:
        mdn.append(mode(df[name].values).mode[0])
        nulls.append(df[name].isnull().sum() / df[name].shape[0])
    df_res['mode'] = mdn
    df_res['part_missing'] = nulls
    df_res.index = numeric
    return df_res
```

```python
def change_left(inter):
    return pd.Interval(left=-np.inf, right=inter.right, closed=inter.closed)
def change_right(inter):
    return pd.Interval(left=inter.left, right=np.inf, closed=inter.closed)
def open_lf_rg(df, feature):
    sortd = np.sort(df[feature].unique())
    df[feature].replace(sortd[0], change_left(sortd[0]), inplace=True)
    df[feature].replace(sortd[-1], change_right(sortd[-1]), inplace=True)
```

```python
def my_woe(df, feature):
    res = pd.DataFrame()
    vals = list()
    woes = list()
    ivs = list()
    all_bad = df['target'].sum()
    all_good = df['target'].shape[0] - all_bad
    #print(all_good, all_bad, df['target'].shape)
    for val in df[feature].unique().tolist():
        vals.append(val)
        bad = df[df[feature] == val]['target'].sum()
        good = df[df[feature] == val]['target'].shape[0] - bad
        w = np.log((good / all_good) / (bad / all_bad))
        if w in [np.inf, -np.inf]:
            print('replacing')
            w = 0
        woes.append(w)
        ivs.append(((good / all_good) - (bad / all_bad)) * w)
    res = pd.DataFrame()
    res['vals'] = vals
    res['woes'] = woes
    res['ivs'] = ivs
    return res
```

```python
def calculate_woe_iv(dataset, feature, target):
    lst = []
    for i in range(dataset[feature].nunique()):
        val = list(dataset[feature].unique())[i]
        lst.append({
            'Value': val,
            'All': dataset[dataset[feature] == val].count()[feature],
            'Good': dataset[(dataset[feature] == val) & (dataset[target] == 0)].count()[feature],
            'Bad': dataset[(dataset[feature] == val) & (dataset[target] == 1)].count()[feature]
        })
        
    dset = pd.DataFrame(lst)
    dset['Distr_Good'] = dset['Good'] / dset['Good'].sum()
    dset['Distr_Bad'] = dset['Bad'] / dset['Bad'].sum()
    dset['WoE'] = np.log(dset['Distr_Good'] / dset['Distr_Bad'])
    dset = dset.replace({'WoE': {np.inf: 0, -np.inf: 0}})
    dset['IV'] = (dset['Distr_Good'] - dset['Distr_Bad']) * dset['WoE']
    iv = dset['IV'].sum()
    
    dset = dset.sort_values(by='WoE')
    
    return dset, iv
#calculate_woe_iv(accept_woe, 'age_bin', 'target')
```

```python
def unite_woes(row, tol):
    result = list()
    is_used = set()
    while True:
        for el in row:
            if el[0] in is_used:
                    continue
            is_used.add(el[0])
            to_unite = set()
            to_unite.add(el[0])
            for sup in row:
                if sup[0] in is_used:
                    continue
                if ((el[1] >= 0) and (sup[1] >= 0)) or ((el[1] < 0) and (sup[1] < 0)):
                    if abs(el[1] - sup[1]) <= tol:
                        to_unite.add(sup[0])
                        is_used.add(sup[0])
            result.append(to_unite)
#         print(tol)
#         print(len(result))
        if len(result) > 5:
            tol += 0.01
            result=list()
            is_used=set()
        else:
            break
    return result


def merge_sets(df, feature, sets):
    df[feature + '_init'] = df[feature].values.tolist()
    df[feature + 'groups'] = df[feature].values.tolist()
    for i, st in enumerate(sets):
        sup = sorted(list(st))
        if len(sup) >= 2:
            df[feature] = [sup if x in sup else x for x in df[feature]]
            df[feature + 'groups'] = ['group' + str(i) if x in sup else x for x in df[feature + 'groups']]
```

```python
def transform_numeric(df, numeric):
    new_features = list()
    grouped = list()
    for ft in numeric:
        bin_size = min(50, df[ft].unique().shape[0])
        print('Processing %s with bin_size %d' % (ft, bin_size))
        ft_bin = ft + '_bin'
        new_features.append(ft_bin)
        df[ft_bin] = pd.qcut(df[ft], bin_size, duplicates='drop')
        open_lf_rg(df, ft_bin)
        kek = my_woe(df, ft_bin)
        #print(kek)
        if bin_size > 10:
            sup = unite_woes(kek.sort_values(by='woes').values.tolist(), 0.02)
            merge_sets(df, ft_bin, sup)
            grouped.append(ft_bin + 'groups')
        else:
            df[ft_bin + '_init'] = df[ft_bin].values.tolist()
            grouped.append(ft_bin)
    return new_features, grouped

new_bins, new_groups = transform_numeric(accept_woe, numeric)
```

```python
def get_IV(df):
    ivs = dict()
    names = df.columns.values.tolist()
    names.remove('target')
    all_bad = df['target'].sum()
    all_good = df['target'].shape[0] - all_bad
    for name in names:
        support = list()
        for el in df[name].unique().tolist():
            #print(name, ' : ', el)
            #tmp = df[df[name] == el]
            bad = df[df[name] == el]['target'].sum()
            good = df[df[name] == el]['target'].shape[0] - bad
            support.append(((good / all_good) - (bad / all_bad)) * el)
        #print(np.sum(np.array(support)))
        ivs[name] = np.sum(np.array(support))
    return pd.DataFrame(ivs, index=['IV']).transpose()

iv_features = get_IV(final_accept)
```
